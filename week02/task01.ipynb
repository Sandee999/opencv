### Imports

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import skimage.data as skdata
import shutil
```

### Save camera.png

```python
camera_img = skdata.camera()
print(f"Size: {camera_img.shape[1]}x{camera_img.shape[0]}")
cv.imwrite("output/camera.png", camera_img)

# Show Image
plt.figure(figsize=(5,5))
plt.imshow(cv.cvtColor(camera_img, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

### Save coins.png

```python
coins_img = skdata.coins()
print(f"Size: {coins_img.shape[1]}x{coins_img.shape[0]}")
cv.imwrite("output/coins.png", coins_img)

# Show Image
plt.figure(figsize=(5,5))
plt.imshow(cv.cvtColor(coins_img, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
