## Thresholding 
- Thresholding is an image segmentation technique that converts a grayscale image into a binary (black-and-white) image by comparing each pixel's intensity with a threshold value.

### Import

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Inputs

```python
img = cv.imread("output/camera.png", cv.IMREAD_GRAYSCALE)
images = []
thresholds = [80, 120, 160, 200]
titles = [
    "Threshold=80",
    "Threshold=120",
    "Threshold=160",
    "Threshold=200",
    "Adaptive Gaussian"
]
```

```python
for t in thresholds:
    _, binary = cv.threshold(img, t, 255, cv.THRESH_BINARY)
    images.append(binary)
```

```python
adaptive = cv.adaptiveThreshold(
    img,
    255,
    cv.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv.THRESH_BINARY,
    21,
    5
)

images.append(adaptive)
```

```python
plt.figure(figsize=(18,4))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.savefig("output/3_threshold.png", dpi=300)
plt.show()
```
