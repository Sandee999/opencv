## Canny Edge Detection

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread("input/personal_edge.jpg", cv.IMREAD_GRAYSCALE)
img2 = cv.imread("output/camera.png", cv.IMREAD_GRAYSCALE)
img3 = cv.imread("output/coins.png", cv.IMREAD_GRAYSCALE)
```

```python
thresholds = [
    (30, 100),
    (50, 150),
    (80, 200),
    (100, 300),
    (150, 400)
]
```

### apertureSize=4

```python
results = []

for low, high in thresholds:
    edges1 = cv.Canny(img1,low,high,apertureSize=5)
    edges2 = cv.Canny(img2,low,high,apertureSize=5)
    edges3 = cv.Canny(img3,low,high,apertureSize=5)
    results.append((edges1, edges2, edges3))
```

```python
plt.figure(figsize=(15, 8))

for i in range(5):
    plt.subplot(3, 5, i+1)
    plt.imshow(results[i][0], cmap='gray')
    plt.title(f"personal_edge.jpg: {thresholds[i]}")
    plt.axis('off')

    plt.subplot(3, 5, 5+i+1)
    plt.imshow(results[i][1], cmap='gray')
    plt.title(f"camera.png: {thresholds[i]}")
    plt.axis('off')
    
    plt.subplot(3, 5, 10+i+1)
    plt.imshow(results[i][2], cmap='gray')
    plt.title(f"coins.png: {thresholds[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
