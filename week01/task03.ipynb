### Imports

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Image artithmetic
- The operation add and subtract allow to add two images. If we add the value (50,50,50) to each pixel, the image becomes brighter - as shown below:

```python
img = cv.imread("input/chelsea.jpg")
M = np.ones(img.shape, dtype='uint8') * 50
bright_img = cv.add(img, M)

# Show Image
plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(bright_img, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

### Save Image

```python
cv.imwrite("output/6_brighter.png", bright_img)
```

### Numeric check

```python
print("Original mean intensity :", img.mean())
print("Brighter mean intensity:", bright_img.mean())
print("Maximum pixel value:", bright_img.max())

assert bright_img.mean() > img.mean()
assert bright_img.max() <= 255

print("All checks passed!")
```
